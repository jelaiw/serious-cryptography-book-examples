Listing 9-3: Generate an insecure RSA private key using the OpenSSL toolkit.

```bash
openssl version
```

```bash
openssl genrsa --help
```

Note, OpenSSL now warns for RSA keys as short as 31 bits.

```bash
openssl genrsa 31
```

```bash
openssl genrsa 384
```

```bash
openssl genrsa 512
```

Looks like the updated warning threshold for this version of OpenSSL is somewhere between 384 and 512 bits.
