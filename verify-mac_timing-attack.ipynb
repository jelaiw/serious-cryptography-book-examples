```python
from time import time
```

```python
def compare_mac(x, y, n):
    for i in range(n):
        if x[i] != y[i]:
            return False
    return True
```

```python
MAC1 = "0123456789abcdef"
MAC2 = "01X3456789abcdef"
TRIALS = 99999
```

```python
# Note, all sixteen bytes are compared.
start = time()
for i in range(TRIALS):
    compare_mac(MAC1, MAC1, len(MAC1))
end = time()
print(f"Elapsed time: {end - start:.2f}")
```

```python
# Note, only 3 bytes are compared.
start = time()
for i in range(TRIALS):
    compare_mac(MAC1, MAC2, len(MAC2))
end = time()
print(f"Elapsed time: {end - start:.2f}")
```
